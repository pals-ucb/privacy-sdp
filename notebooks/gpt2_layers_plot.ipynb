```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure
fig, ax = plt.subplots(figsize=(12, 8))

# Draw the Embeddings Layer
embeddings_box = patches.FancyBboxPatch((0.1, 0.7), 0.3, 0.1,
                                        boxstyle="round,pad=0.02",
                                        edgecolor="black", facecolor="lightblue",
                                        linewidth=2)
ax.add_patch(embeddings_box)
plt.text(0.25, 0.75, "Embeddings Layer\n(25M params)", ha="center", fontsize=12)

# Draw the Hidden Layer (single instance)
hidden_box = patches.FancyBboxPatch((0.1, 0.4), 0.3, 0.2,
                                    boxstyle="round,pad=0.02",
                                    edgecolor="black", facecolor="lightgreen",
                                    linewidth=2)
ax.add_patch(hidden_box)
plt.text(0.25, 0.55, "Hidden Layer (1 of 12)\n(50M params)", ha="center", fontsize=12)

# Draw the attention sub-layer
attention_box = patches.FancyBboxPatch((0.15, 0.45), 0.2, 0.05,
                                       boxstyle="round,pad=0.02",
                                       edgecolor="black", facecolor="lightyellow",
                                       linewidth=2)
ax.add_patch(attention_box)
plt.text(0.25, 0.465, "Attention\nSub-layer", ha="center", fontsize=10)

# Draw the Feed-Forward sub-layer
ff_box = patches.FancyBboxPatch((0.15, 0.375), 0.2, 0.05,
                                boxstyle="round,pad=0.02",
                                edgecolor="black", facecolor="lightyellow",
                                linewidth=2)
ax.add_patch(ff_box)
plt.text(0.25, 0.39, "Feed-Forward\nSub-layer", ha="center", fontsize=10)

# Draw the Output Layer
output_box = patches.FancyBboxPatch((0.1, 0.1), 0.3, 0.1,
                                    boxstyle="round,pad=0.02",
                                    edgecolor="black", facecolor="lightcoral",
                                    linewidth=2)
ax.add_patch(output_box)
plt.text(0.25, 0.15, "Output Layer\n(20M params)", ha="center", fontsize=12)

# Draw multiplicity factors
plt.arrow(0.425, 0.55, 0, 0.1, head_width=0.025, head_length=0.02, fc='k', ec='k')
plt.text(0.45, 0.6, "x 12", ha="center", fontsize=12)

# Add arrows between layers
plt.arrow(0.25, 0.68, 0, -0.04, head_width=0.025, head_length=0.02, fc='k', ec='k')
plt.arrow(0.25, 0.35, 0, -0.1, head_width=0.025, head_length=0.02, fc='k', ec='k')

# Setting plot limits and removing axes
plt.xlim(0, 1)
plt.ylim(0, 1)
ax.axis('off')

# Title
plt.title("GPT-2 Neural Network Key Layers", fontsize=16)

# Save the figure
plt.savefig("GPT2_Key_Layers.png")
plt.show()
```
